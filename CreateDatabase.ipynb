# Create Database

Supply the URL to the CSV version of your publisher and journal info to build a SQLite version of that information.



```python
PUB_URL = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRPFcOiXgZAo6XENgRXj3FoQ_BnbsYtAJq0QtlnhjGmpIkQjIp8eZNX6C66tcaooh1pfaUR8AULfSji/pub?gid=1143475194&single=true&output=csv"
JOURNAL_URL = "https://docs.google.com/spreadsheets/d/e/2PACX-1vRPFcOiXgZAo6XENgRXj3FoQ_BnbsYtAJq0QtlnhjGmpIkQjIp8eZNX6C66tcaooh1pfaUR8AULfSji/pub?gid=0&single=true&output=csv"

sqliteDBName = "src/apc-details.db"
```

```python
import pandas as pd
import sqlite3
```

```python
journalDF = pd.read_csv(JOURNAL_URL)
pubDF = pd.read_csv(PUB_URL)
combinedDF = pd.merge(journalDF,pubDF, on="Publisher")
```

```python
combinedDF
```

```python
conn = sqlite3.connect(sqliteDBName)
pubDF.to_sql('publisherInfo',conn, if_exists='replace',index=False)
journalDF.to_sql('journalInfo',conn, if_exists='replace',index=False)
combinedDF.to_sql('apcDetails', conn, if_exists='replace', index=False)
conn.close()
print("fin")
```
